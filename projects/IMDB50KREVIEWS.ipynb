```python
import re
def separate_contractions(text):
    if not isinstance(text, str):
        return text
    # Separate common contractions stuck to previous word
    text = re.sub(
        r'([a-zA-Z])('
        r"n't|'ll|'re|'ve|'d|'m|'s"
        r')\b',
        r'\1 \2',
        text
    )

    return text
```

```python
import re

def separate_after_contractions(text):
    return re.sub(
        r"(n't|'ll|'re|'ve|'d|'m|'s)([A-Za-z])",
        r"\1 \2",
        text
    )
```

```python
import contractions

df['review'] = df['review'].astype(str).apply(separate_after_contractions).apply(separate_contractions).apply(contractions.fix)
```

#expanding abbreviations


```python
ABBREVIATIONS = {
    "u": "you",
    "ur": "your",
    "idk": "i do not know",
    "imo": "in my opinion",
    "imho": "in my humble opinion",
    "btw": "by the way",
    "thx": "thanks",
    "pls": "please",
    "plz": "please",
    "asap": "as soon as possible",
    "bc": "because",
    "bcz": "because",
    "cuz": "because",
    "coz": "because",
    "msg": "message",
    "tmrw": "tomorrow",
    "w/": "with",
    "w/o": "without",
    "omg": "oh my god",
    "lol": "laugh out loud",
    "lmao": "laughing my ass off",
    "rofl": "rolling on the floor laughing",
    "tbh": "to be honest",
    "fyi": "for your information",
    "afaik": "as far as i know"
}
```

```python
import re

def expand_abbreviations(text):
    if not isinstance(text, str):  # This line returns as soon as the input is non string
        return text

    pattern = r'\b(' + '|'.join(map(re.escape, ABBREVIATIONS.keys())) + r')\b'  # This line builds a regular expression dynamically from your abbreviation dictionary

    return re.sub(
        pattern,
        lambda m: ABBREVIATIONS[m.group().lower()],
        text,
        flags=re.IGNORECASE
    )
```

```python
df['review'] = df['review'].apply(expand_abbreviations)
```
