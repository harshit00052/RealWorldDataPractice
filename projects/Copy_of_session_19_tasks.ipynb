### `Q-4:` Perform `groupby` based on the `"Team"` column and then perform Z Normalization on top of the below columns of each group:
1. Passes
2. Passes Completed
3. Attempted Line Breaks
4. Completed Line Breaks

You have to make a python function named `z_normalization` which takes two arguments:

1. *group:* Every group that you have created
2. *cols_to_perform:* This parameter takes a list of columns on which you have to perform the Z-Normalization.

For this problem, you have to use th `apply()` method.

$\Large Z \ - \ Normalization = \frac{X_i - \mu}{std} $

After that find out the below values for each group:
- minimum "Passess"
- maximum "Passess"
- minimum "Yellow Cards"
- maximum "Yellow cards"
- average "Yellow Cards"
- maximum "Attempted Line Breaks"
- minimum "Attempted Line Breaks"
- standardard deviation of "Attempted Line Breaks"
- average Possession

````python
def z_normalization(group, cols_to_perform):
  tcol = group[cols_to_perform]
  return (tcol-tcol.mean())/tcol.std();
````

````python
norm_df = fifa_df.groupby('Team').apply(z_normalization, cols_to_perform = ['Passes', 'Passes Completed', 'Attempted Line Breaks', 'Completed Line Breaks'])
````

````python
norm_df
````

````python
temp_group = fifa_df.groupby('Team')
for group, tdf in temp_group:
  print("Team",group)
  print("minimum Passess ",tdf['Passes'].min())
  print("maximum Passess ",tdf['Passes'].max())
  print("average Yellow Cards",tdf['Yellow Cards'].mean().round(2))
  print('\n')
````

###`Q-5:` Find batsman in below category-
* Highest score while chasing
* Best Strike rate while chasing and have faced 100+ balls


> Chasing mean team batting in second inning

````python
ipl_df[ipl_df['innings']==2].groupby('batter').agg({'total_run':'sum'}).reset_index().sort_values('total_run', ascending=False).head(1)
````

````python
def find_strike_rate(group, cols):
  g_col = group[cols]
  balls = len(g_col)
  if(balls>100):
    return (g_col.sum()/balls)*100

ipl_df.groupby('batter').apply(find_strike_rate, cols='total_run', include_groups=False).dropna().sort_values(ascending=False)
````

````python
temp_ipl_df = ipl_df.groupby('batter').agg({'total_run':['sum', 'count']}).reset_index()
temp_ipl_df = temp_ipl_df[temp_ipl_df['total_run']['count']>100]
temp_ipl_df['strike_rate'] = (temp_ipl_df['total_run']['sum']/temp_ipl_df['total_run']['count'])*100
temp_ipl_df.sort_values(by='strike_rate',ascending=False)
````
